# Boxplot 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
df = pd.read_csv(url, sep = "|")
df = df.dropna()
df.head()
```

```python
sns.boxplot(df["age"]);
```

```python
sns.boxplot(y = df["age"]);
```

```python
sns.set_style("whitegrid")
sns.boxplot(y = df["age"]);
```

```python
plt.figure(figsize = (3,5))
sns.boxplot(y = df["age"]);
```

```python
plt.figure(figsize = (3,5))
sns.boxplot(y = df["age"])
plt.title("Yaş Yoğunluğu")
plt.ylabel("Yaş");
```

```python
plt.figure(figsize = (5,5))
sns.boxplot(x = df["gender"], y = df["age"]);
```

```python
plt.figure(figsize = (5,5))
sns.boxplot(x = df["gender"], y = df["age"]);
plt.title("Cinsiyetlerin Yaş Yoğunlu")
plt.xlabel("Cinsiyet")
plt.ylabel("Yaş");
```

```python
plt.figure(figsize = (5,5))
sns.boxplot(x = df["gender"], y = df["age"], palette = "Set1")
plt.title("Cinsiyetlerin Yaş Yoğunlu")
plt.xlabel("Cinsiyet")
plt.ylabel("Yaş");
```

```python
plt.figure(figsize = (5,5))
colors = {"M" : "seagreen", "F" : "blueviolet"}
sns.boxplot(x = df["gender"], y = df["age"], palette = colors)
plt.title("Cinsiyetlerin Yaş Yoğunlu")
plt.xlabel("Cinsiyet")
plt.ylabel("Yaş");
```

```python
plt.figure(figsize = (5,5))
colors = {"M" : "seagreen", "F" : "blueviolet"}
sns.boxplot(x = df["gender"], y = df["age"], palette = colors, linewidth = 5)
plt.title("Cinsiyetlerin Yaş Yoğunlu")
plt.xlabel("Cinsiyet")
plt.ylabel("Yaş");
```

```python
plt.figure(figsize = (5,5))
colors = {"M" : "seagreen", "F" : "blueviolet"}
sns.boxplot(x = df["gender"], y = df["age"], palette = colors, width = 0.3)
plt.title("Cinsiyetlerin Yaş Yoğunlu")
plt.xlabel("Cinsiyet")
plt.ylabel("Yaş");
```

```python
plt.figure(figsize = (5,5))
colors = {"M" : "seagreen", "F" : "blueviolet"}
sns.boxplot(x = df["gender"], y = df["age"], palette = colors, notch = True)
plt.title("Cinsiyetlerin Yaş Yoğunlu")
plt.xlabel("Cinsiyet")
plt.ylabel("Yaş");
```

```python
plt.figure(figsize = (5,5))
colors = {"M" : "seagreen", "F" : "blueviolet"}
sns.boxplot(x = df["gender"], y = df["age"], palette = colors, linewidth = 5, notch = True)
plt.title("Cinsiyetlerin Yaş Yoğunlu")
plt.xlabel("Cinsiyet")
plt.ylabel("Yaş");
```

```python
plt.figure(figsize = (5,5))
colors = {"M" : "seagreen", "F" : "blueviolet"}
sns.boxplot(x = df["gender"], y = df["age"], palette = colors, linewidth = 5, width=0.3, notch = True)
plt.title("Cinsiyetlerin Yaş Yoğunlu")
plt.xlabel("Cinsiyet")
plt.ylabel("Yaş");
```

```python
plt.figure(figsize = (5,5))
colors = {"M" : "seagreen", "F" : "blueviolet"}
sns.stripplot(x = df["gender"], y = df["age"], jitter = True, palette = "Greys",  linewidth = 1 , size = 5)
sns.boxplot(x = df["gender"], y = df["age"], palette = colors)
plt.title("Cinsiyetlerin Yaş Yoğunlu")
plt.xlabel("Cinsiyet")
plt.ylabel("Yaş");
```

```python
plt.figure(figsize = (5,5))
colors = {"M" : "seagreen", "F" : "blueviolet"}
sns.stripplot(x = df["gender"], y = df["age"], jitter = True, palette = "Greys",  linewidth = 1 , size = 7, marker = "D")
sns.boxplot(x = df["gender"], y = df["age"], palette = colors)
plt.title("Cinsiyetlerin Yaş Yoğunlu")
plt.xlabel("Cinsiyet")
plt.ylabel("Yaş");
```

```python
ordered = df.groupby(by=["gender"])["age"].median().sort_values().index
median = df.groupby(["gender"])["age"].median().values
nobs = df["gender"].value_counts().values
nobs = [str(x) for x in nobs.tolist()]
nobs = ["n:" + i for i in nobs]
pos = range(len(nobs))

plt.figure(figsize = (5,5))
colors = {"M" : "seagreen", "F" : "blueviolet"}
ax = sns.boxplot(x = df["gender"], y = df["age"], palette = colors)
plt.title("Cinsiyetlerin Yaş Yoğunlu")
plt.xlabel("Cinsiyet")
plt.ylabel("Yaş")
for tick, label in zip(pos, ax.get_xticklabels()):
    ax.text(pos[tick], median[tick]+2, nobs[tick], horizontalalignment = "center", size = "large", color = "w", weight = "semibold",
            style = "italic", family = "monospace")
```

```python
plt.figure(figsize = (20,5))
plt.xticks(rotation = 45)
sns.boxplot(x = df["occupation"], y = df["age"], palette = "twilight")
plt.title("MESLEKLERİN YAŞ YOĞUNLUĞU")
plt.xlabel("MESLEK")
plt.ylabel("YAŞ");
```

```python
colors = {occupation: "lightseagreen" if occupation == "educator" else "grey" for occupation in df.occupation.unique()}
plt.figure(figsize = (20,5))
plt.xticks(rotation = 45)
sns.boxplot(x = df["occupation"], y = df["age"], palette = colors)
plt.title("MESLEKLERİN YAŞ YOĞUNLUĞU")
plt.xlabel("MESLEK")
plt.ylabel("YAŞ");
```

```python
plt.figure(figsize = (20,5))
plt.xticks(rotation = 45)
sns.boxplot(x = df["occupation"], y = df["age"], hue = df["gender"], palette = "cool")
plt.title("MESLEKLERİN YAŞ YOĞUNLUĞU")
plt.xlabel("MESLEK")
plt.ylabel("YAŞ");
```

```python
ordered = df.groupby(by=["occupation"])["age"].median().sort_values().index
plt.figure(figsize = (20,5))
plt.xticks(rotation = 45)
sns.boxplot(x = df["occupation"], y = df["age"], palette = "viridis", order = ordered)
plt.title("MESLEKLERİN YAŞ YOĞUNLUĞU")
plt.xlabel("MESLEK")
plt.ylabel("YAŞ");
```
