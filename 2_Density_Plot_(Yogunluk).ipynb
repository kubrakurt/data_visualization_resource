# Density Plot (Yoğunluk Grafiği)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
df = pd.read_csv(url, sep = "|")
df = df.dropna()
df.head()
```

```python
sns.kdeplot(df["age"]);
```

```python
sns.kdeplot(df["age"], shade = True);
```

```python
sns.set_style("whitegrid")
```

```python
sns.set_style("whitegrid")
sns.kdeplot(df["age"], shade = True);
```

```python
sns.set_style("darkgrid")
sns.kdeplot(df["age"], shade = True);
```

```python
sns.kdeplot(df["age"], color = "red");
```

```python
sns.kdeplot(df["age"], color = "red", shade = True);
```

```python
sns.kdeplot(df["age"], color = "red", shade = True, vertical = True);
```

```python
plt.figure(figsize = (4,6))
sns.kdeplot(df["age"], color = "red", shade = True, vertical = True);
```

```python
sns.kdeplot(df["age"], color = "palevioletred", shade = True)
plt.title("Yaş Yoğunluk Grafiği");
```

```python
sns.kdeplot(df["age"], color = "palevioletred", shade = True)
plt.title("Yaş Yoğunluk Grafiği")
plt.gca().legend_ = None;
```

```python
sns.kdeplot(df["age"], color = "palevioletred", shade = True, cumulative = True)
plt.title("Yaş Yoğunluk Grafiği")
plt.gca().legend_ = None;
```

```python
sns.kdeplot(df["age"], color = "palevioletred", shade = True, bw = 0.5)
plt.title("Yaş Yoğunluk Grafiği")
plt.gca().legend_ = None;
```

```python
sns.kdeplot(df["age"], color = "palevioletred", shade = True, bw = 0.05)
plt.title("Yaş Yoğunluk Grafiği")
plt.gca().legend_ = None;
```

# [Data](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv)

```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv"
df = pd.read_csv(url)
df = df.dropna(axis = 1)
df.head()
```

```python
df.info()
```

```python
sns.set_style("whitegrid")
sns.kdeplot(df["mpg"], color = "mediumpurple")
sns.kdeplot(df["acceleration"], color = "gold");
```

```python
sns.kdeplot(df["mpg"], color = "mediumpurple", shade = True)
sns.kdeplot(df["acceleration"], color = "gold", shade = True);
```

```python
plt.figure(figsize = (10,5))
sns.kdeplot(df["mpg"], color = "mediumpurple", shade = True)
sns.kdeplot(df["acceleration"], color = "gold", shade = True);
```

```python
sns.kdeplot(df["mpg"], color = "mediumpurple", shade = True, cumulative = True)
sns.kdeplot(df["acceleration"], color = "gold", shade = True);
```

```python
sns.kdeplot(df["mpg"], color = "mediumpurple", shade = True)
sns.kdeplot(df["acceleration"], color = "gold", shade = True, cumulative = True);
```

```python
sns.kdeplot(df["mpg"], color = "mediumpurple", shade = True, cumulative = True)
sns.kdeplot(df["acceleration"], color = "gold", shade = True, cumulative = True);
```

```python
sns.kdeplot(df["mpg"], color = "mediumpurple", shade = True)
sns.kdeplot(df["acceleration"], color = "gold", shade = True)
plt.title("mpg & acceleration density plot");
```

```python
sns.kdeplot(df["mpg"], color = "mediumpurple", shade = True, bw = 0.5)
sns.kdeplot(df["acceleration"], color = "gold", shade = True, bw = 0.5)
plt.title("mpg & acceleration density plot");
```

```python
df["horsepower"] = pd.to_numeric(df["horsepower"], errors = "coerce")
sns.kdeplot(df["mpg"], color = "mediumpurple", shade = True, bw = 0.05)
sns.kdeplot(df["acceleration"], color = "gold", shade = True, bw = 0.05)
plt.title("mpg & acceleration density plot");
```

```python
sns.kdeplot(df["mpg"], color = "mediumpurple", shade = True)
sns.kdeplot(df["acceleration"], color = "gold", shade = True)
sns.kdeplot(df["horsepower"], color = "lightseagreen", shade = True)
plt.title("mpg & acceleration density plot");
```
